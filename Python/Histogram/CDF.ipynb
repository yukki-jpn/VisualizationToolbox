## 一般のCDFの書き方
CDF ($F(x)$) は、下記のようにして定義される。
$$
\begin{align}
F(x^*) = P(x^*>x)= \sum_{x<x^*}p(x)
\end{align}
$$
Pythonでは、seaborn のecdfplotかmatplotlib.pyplotのhistを使用して書く。

```python
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

x = stats.expon.rvs(scale=1,size=10000)
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,5))
sns.ecdfplot(x,ax=ax1)
ax1.set_xlabel(r"$x$",fontsize=15)
ax1.set_ylabel(r"$P_{>}(x)$",fontsize=15)
ax1.set_facecolor('0.9')

ax2.hist(x,cumulative=1,bins=100,density=True,histtype="step")
ax2.set_xlabel(r"$x$",fontsize=15)
ax2.set_ylabel(r"$P_{>}(x)$",fontsize=15)
ax2.set_facecolor('0.9')

ax1.set_title("seaborn")
ax2.set_title("matplotlib")
```

### 信頼区間
CDFの推定方法（定義）を見てわかるように、CDFのテールの推定に使用されるデータ数はヘッドの推定に使用されるデータ数と比較してずっと小さい。つまり、CDFの信頼区間のようなものを考えたとき、その信頼区間はヘッドからテールにかけて扇形に広がっていくと考えられる。テールの推定値の妥当性は、各個人が目で見て確かめる必要がある: 下の図の数値シミュレーションの結果によると、CDF信頼区間はテールに行くにつれて広がっていくことが分かる。

```python
# イメージ図
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

x = stats.expon.rvs(scale=1,size=10000)
fig,ax=plt.subplots(figsize=(5,5))
sns.ecdfplot(x,complementary=True,ax=ax,color="blue",zorder=101,label="Empirical CDF")
for i in range(100):
    X = stats.expon.rvs(scale=1,size=10000)
    sns.ecdfplot(X,complementary=True,ax=ax,color="lightblue",alpha=0.6)
    pass
ax.set_xlabel(r"$x$",fontsize=15)
ax.set_ylabel(r"$P_{>}(x)$",fontsize=15)
ax.set_title("Numerical confidential interval of CDF",fontsize=15)
ax.set_facecolor('0.9')
ax.set_yscale("log")
```
